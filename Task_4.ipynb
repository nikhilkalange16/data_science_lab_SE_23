# Task
Set up the environment, perform statistical analysis (e.g., descriptive statistics) on a sample dataset, conduct a probability simulation (e.g., coin flips or dice rolls), visualize the results using Matplotlib with legends, and finally summarize the findings.

## Set up environment

### Subtask:
Import necessary Python libraries: NumPy for numerical operations, SciPy for advanced statistics, and Matplotlib for visualization.


**Reasoning**:
The subtask is to import specific Python libraries: NumPy, SciPy (stats module), and Matplotlib (pyplot module). I will use a code block to import these libraries with their standard aliases.



```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt

print("Libraries imported successfully: numpy (as np), scipy.stats, and matplotlib.pyplot (as plt).")
```

## Perform Statistical Analysis

### Subtask:
Demonstrate statistical analysis, such as calculating descriptive statistics (mean, median, standard deviation) on a sample dataset using NumPy and SciPy.


**Reasoning**:
I need to generate a sample dataset and then calculate its mean, median, and standard deviation using NumPy, as requested in the instructions.



```python
np.random.seed(42) # for reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=1000)

mean_val = np.mean(sample_data)
median_val = np.median(sample_data)
std_dev_val = np.std(sample_data)

print(f"Sample data generated with 1000 points (mean=50, std=10).")
print(f"Calculated Mean: {mean_val:.2f}")
print(f"Calculated Median: {median_val:.2f}")
print(f"Calculated Standard Deviation: {std_dev_val:.2f}")
```

## Run Probability Simulation

### Subtask:
Conduct a probability simulation, for example, simulating a large number of coin flips or dice rolls, and analyze the outcomes.


I will simulate a series of coin flips to demonstrate a probability simulation. I'll define the number of flips, use `np.random.randint` to generate the outcomes (0 for tails, 1 for heads), then count the occurrences of heads and tails, and finally calculate their proportions. The results will be printed to show the analysis of the simulation.

**Reasoning**:
I need to implement the coin flip simulation as described in the instructions, which includes defining the number of flips, simulating them, counting heads and tails, calculating proportions, and printing the results.



```python
num_flips = 10000

# Simulate coin flips (0 for tails, 1 for heads)
# np.random.randint(low, high, size) generates integers from low (inclusive) to high (exclusive).
# So, randint(0, 2, num_flips) will generate 0s and 1s.
coin_flips = np.random.randint(0, 2, num_flips)

# Count heads (1s) and tails (0s)
heads_count = np.sum(coin_flips == 1)
tails_count = np.sum(coin_flips == 0)

# Calculate proportions
heads_proportion = heads_count / num_flips
tails_proportion = tails_count / num_flips

print(f"--- Coin Flip Simulation Results ---")
print(f"Total Number of Flips: {num_flips}")
print(f"Number of Heads: {heads_count}")
print(f"Number of Tails: {tails_count}")
print(f"Proportion of Heads: {heads_proportion:.4f}")
print(f"Proportion of Tails: {tails_proportion:.4f}")
```

## Visualize Results

### Subtask:
Generate plots using Matplotlib to visualize the results of the statistical analysis and the distribution of outcomes from the probability simulation. Ensure legends are provided for clarity.


**Reasoning**:
I need to generate two plots side-by-side using Matplotlib to visualize the sample data distribution and the coin flip simulation results, including legends and titles as per the instructions.



```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Histogram of Sample Data with Mean and Median
axes[0].hist(sample_data, bins=30, alpha=0.7, color='skyblue', label='Sample Data Distribution')
axes[0].axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}')
axes[0].axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_val:.2f}')
axes[0].set_title('Distribution of Sample Data with Mean and Median')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Plot 2: Coin Flip Simulation Outcomes
outcomes = ['Heads', 'Tails']
proportions = [heads_proportion, tails_proportion]
axes[1].bar(outcomes, proportions, color=['lightcoral', 'lightgreen'])
axes[1].set_title('Coin Flip Simulation Outcomes')
axes[1].set_xlabel('Outcome')
axes[1].set_ylabel('Proportion')

# Add main title to the figure
fig.suptitle('Statistical Analysis and Probability Simulation Results', fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make space for suptitle
plt.show()
print("Visualizations generated successfully.")
```
