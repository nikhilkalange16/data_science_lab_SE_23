```python
def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = 7
print(f"The factorial of {number} is {factorial(number)}")
```

### 3. Program to identify if the entered number is palindrome or not

```python
def check_palindrome(number):
    # Convert number to string to easily reverse it
    num_str = str(number)
    reversed_num_str = num_str[::-1]

    if num_str == reversed_num_str:
        print(f"{number} is a palindrome.")
    else:
        print(f"{number} is not a palindrome.")

# Example usage:
check_palindrome(121)
check_palindrome(12321)
check_palindrome(12345)
```

#### 2. Implement Breadth-First Search (BFS)

BFS explores a graph layer by layer, visiting all neighbors at the current depth before moving on to the nodes at the next depth level. It typically uses a queue data structure to keep track of the nodes to visit.


**Reasoning**:
Now, I will implement the Breadth-First Search (BFS) algorithm in a Python code block, as introduced in the previous markdown block, using the defined graph and a starting node 'A'.



```python
from collections import deque

def bfs(graph, start_node):
    visited = set()
    queue = deque([start_node])
    bfs_traversal = []

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            bfs_traversal.append(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
    return bfs_traversal

print("BFS Traversal starting from node 'A':")
b_traversal = bfs(graph, 'A')
print(b_traversal)
```

#### 3. Implement Depth-First Search (DFS)

DFS explores a graph by going as deep as possible along each branch before backtracking. It typically uses a stack data structure (or recursion, which implicitly uses the call stack) to keep track of the nodes to visit.

**Reasoning**:
Now, I will implement the Depth-First Search (DFS) algorithm in a Python code block, as introduced in the previous markdown block, using the defined graph and a starting node 'A'.



```python
def dfs(graph, start_node):
    visited = set()
    stack = [start_node]
    dfs_traversal = []

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            dfs_traversal.append(node)
            # DFS explores depth-first, so push neighbors onto stack
            # in reverse order if you want to process them in alphabetical/given order
            # For simple demonstration, iterating over neighbors and adding them is fine.
            # If explicit order is required (e.g., left-to-right in a tree),
            # one might need to reverse the neighbors list before pushing.
            for neighbor in reversed(graph[node]): # Reversed to maintain consistent order if multiple neighbors
                if neighbor not in visited:
                    stack.append(neighbor)
    return dfs_traversal

print("\nDFS Traversal starting from node 'A':")
d_traversal = dfs(graph, 'A')
print(d_traversal)
```

### 3. Find the Inverse of a Matrix

The inverse of a square matrix `A`, denoted as `A⁻¹`, is a matrix such that when multiplied by `A`, it yields the identity matrix. A matrix is invertible if and only if its determinant is non-zero.

Note: For our example matrix A, the determinant is very close to zero, meaning it's a singular matrix and does not have a true inverse. We'll use a different matrix for the inverse example.

```python
import numpy as np

# Define an invertible 2x2 matrix
B = np.array([[4, 7],
              [2, 6]])

print("Original Matrix B:")
print(B)

det_B = np.linalg.det(B)
print(f"Determinant of Matrix B: {det_B}")

if det_B != 0:
    inv_B = np.linalg.inv(B)
    print("\nInverse of Matrix B:")
    print(inv_B)

    # Verify the inverse (B * B_inv should be close to identity matrix)
    identity_check = np.dot(B, inv_B)
    print("\nVerification (B * B_inv):")
    print(identity_check)
else:
    print("Matrix B is singular and does not have an inverse.")
```

#### 2. Implement A* Search Algorithm

The A* search algorithm is an informed search algorithm that finds the shortest path between two nodes in a graph. It uses a heuristic function to estimate the cost from the current node to the goal node, combining it with the actual cost from the start node to the current node (g-score).

The algorithm works by maintaining a priority queue of nodes to visit, ordered by their f-score (f = g + h), where 'g' is the cost from the start node to the current node, and 'h' is the heuristic estimate from the current node to the goal node.

We'll use Python's `heapq` module to implement the priority queue efficiently.

**Reasoning**:
Now, I will implement the A* search algorithm in a Python code block, using the `heapq` module for the priority queue, as described in the previous markdown block.



```python
import heapq

def astar_search(graph, heuristics, start, goal):
    # Priority queue: (f_score, g_score, current_node, path)
    # f_score = g_score + h_score
    priority_queue = [(heuristics[start], 0, start, [start])]

    # Dictionary to store the minimum g_score found so far for each node
    g_scores = {node: float('inf') for node in graph}
    g_scores[start] = 0

    while priority_queue:
        f_score, g_score, current_node, path = heapq.heappop(priority_queue)

        if current_node == goal:
            return path, g_score

        # If we found a shorter path to current_node already, skip
        if g_score > g_scores[current_node]:
            continue

        for neighbor, weight in graph[current_node]:
            # Calculate tentative g_score for neighbor
            tentative_g_score = g_score + weight

            # If this path to neighbor is better than any previous one
            if tentative_g_score < g_scores[neighbor]:
                g_scores[neighbor] = tentative_g_score
                h_score = heuristics[neighbor]
                f_score = tentative_g_score + h_score
                new_path = path + [neighbor]
                heapq.heappush(priority_queue, (f_score, tentative_g_score, neighbor, new_path))

    return None, float('inf') # No path found

# Example usage:
start_node_astar = 'A'
goal_node_astar = 'G'

path, cost = astar_search(graph_astar, heuristics, start_node_astar, goal_node_astar)

print(f"\nA* Search from {start_node_astar} to {goal_node_astar}:")
if path:
    print(f"Shortest path: {' -> '.join(path)}")
    print(f"Total cost: {cost}")
else:
    print("No path found.")
```

#### 1. Define a Simple Game Tree

To demonstrate the Minimax algorithm, we'll start by defining a simple game tree. This tree will represent possible game states and their outcomes (utility values).

We'll use a dictionary to represent the game tree, where each key is a node (game state) and its value is either a list of its child nodes (if it's an internal node) or an integer representing the utility (score) if it's a leaf node. We'll also indicate which player's turn it is at each internal node (MAX or MIN).

**Reasoning**:
Now, I will define the sample game tree for the Minimax algorithm in a Python code block, as introduced in the previous markdown block.



```python
game_tree = {
    'root': {'A': 'MAX', 'children': ['A', 'B']},
    'A': {'B1': 'MIN', 'children': ['A1', 'A2']},
    'B': {'B2': 'MIN', 'children': ['B1', 'B2']},
    'A1': 3,
    'A2': 12,
    'B1': 8,
    'B2': 2
}

# A more common way to represent for easier Minimax implementation:
# Each node is represented as its utility if it's a leaf, or a list of children if it's an internal node.
# We'll explicitly pass whether it's a maximizing or minimizing player's turn.

# Example Game Tree:
#           MAX
#          /   \
#        MIN     MIN
#       / \     / \
#      3  12   8   2

game_tree_nodes = {
    'root': ['A', 'B'],
    'A': ['A1', 'A2'],
    'B': ['B1', 'B2'],
    'A1': 3, # Leaf node utility
    'A2': 12, # Leaf node utility
    'B1': 8, # Leaf node utility
    'B2': 2  # Leaf node utility
}

print("Game Tree Nodes Defined:")
for node, value in game_tree_nodes.items():
    print(f"{node}: {value}")

# For simplicity in Minimax, we will often define the tree directly as a nested structure or pass player turns explicitly.
# Let's consider a simpler nested representation for the Minimax function for the demonstration.
# Here, we'll assume the root is MAX's turn, then children are MIN's turn, etc.
# (3, 12) -> A's children, (8, 2) -> B's children
# This means 'A' would evaluate to min(3, 12) = 3
# 'B' would evaluate to min(8, 2) = 2
# 'root' would evaluate to max(3, 2) = 3
```

#### 2. Implement the Minimax Algorithm

The Minimax algorithm is a recursive algorithm used in decision-making and game theory, specifically for two-player, zero-sum games. It works by minimizing the possible loss for a worst-case (maximum loss) scenario. When applying it, the player is called 'MAX' and the opponent 'MIN'.

- **Maximizing Player (MAX)**: Tries to get the highest possible score.
- **Minimizing Player (MIN)**: Tries to get the lowest possible score (which is the highest for them if they were MAX).

The algorithm explores the game tree to a certain depth (or until terminal states are reached), assigning utility values to terminal nodes. Then, it backtracks up the tree, calculating the optimal value for each internal node based on whether it's a MAX's turn or a MIN's turn.

We will implement a `minimax` function that takes the current node, the current depth, and a boolean indicating if it's the maximizing player's turn.

**Reasoning**:
Now, I will implement the Minimax algorithm in a Python code block, using the defined `game_tree_nodes` as introduced in the previous markdown block.



```python
def minimax(node_id, depth, maximizing_player, game_tree, max_depth):
    # If the node is a leaf node (has a utility value) or max_depth is reached
    if isinstance(game_tree[node_id], int) or depth == max_depth:
        return game_tree[node_id]

    if maximizing_player:
        best_value = -float('inf')
        for child_id in game_tree[node_id]:
            value = minimax(child_id, depth + 1, False, game_tree, max_depth)
            best_value = max(best_value, value)
        return best_value
    else: # Minimizing player
        best_value = float('inf')
        for child_id in game_tree[node_id]:
            value = minimax(child_id, depth + 1, True, game_tree, max_depth)
            best_value = min(best_value, value)
        return best_value

# Example usage with the defined game_tree_nodes:
# For this simple tree, depth doesn't make much sense as it's a fixed depth tree.
# We'll set max_depth to a value larger than the tree depth for full exploration.

initial_node = 'root'
optimal_value = minimax(initial_node, 0, True, game_tree_nodes, 2) # max_depth 2 for our 2-level tree

print(f"\nMinimax Algorithm Result for the Game Tree:")
print(f"The optimal value for the Maximizing player starting from '{initial_node}' is: {optimal_value}")

# To demonstrate decision making, let's see which path leads to this value from root
def find_optimal_path(node_id, depth, maximizing_player, game_tree, max_depth, target_value):
    if isinstance(game_tree[node_id], int) or depth == max_depth:
        return [node_id] if game_tree[node_id] == target_value else None

    if maximizing_player:
        for child_id in game_tree[node_id]:
            value = minimax(child_id, depth + 1, False, game_tree, max_depth)
            if value == target_value:
                path_from_child = find_optimal_path(child_id, depth + 1, False, game_tree, max_depth, target_value)
                if path_from_child:
                    return [node_id] + path_from_child
    else: # Minimizing player
        for child_id in game_tree[node_id]:
            value = minimax(child_id, depth + 1, True, game_tree, max_depth)
            if value == target_value:
                path_from_child = find_optimal_path(child_id, depth + 1, True, game_tree, max_depth, target_value)
                if path_from_child:
                    return [node_id] + path_from_child
    return None

optimal_path = find_optimal_path(initial_node, 0, True, game_tree_nodes, 2, optimal_value)

if optimal_path:
    print(f"Optimal path (decisions) for the Maximizing player: {' -> '.join(optimal_path)}")
else:
    print("Could not determine optimal path.")
```
