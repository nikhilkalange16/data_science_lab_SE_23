## Implement and Demonstrate BFS

### Subtask:
Implement the Breadth-First Search (BFS) algorithm and demonstrate its traversal on a sample graph. This step will include defining the graph, implementing the BFS function, and printing the traversal order.


**Reasoning**:
The first step in implementing BFS is to define a graph. As per the instructions, I will define a graph using an adjacency list representation, which is a dictionary where keys are nodes and values are lists of their neighbors.



```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

print("Graph defined using adjacency list:")
for node, neighbors in graph.items():
    print(f"{node}: {neighbors}")
```

**Reasoning**:
Now that the graph is defined, I will implement the BFS algorithm as a Python function, following the instructions to keep track of visited nodes and use a queue.



```python
from collections import deque

def bfs(graph, start_node):
    visited = set()
    queue = deque([start_node])
    traversal_order = []

    visited.add(start_node)

    while queue:
        current_node = queue.popleft()
        traversal_order.append(current_node)

        for neighbor in graph[current_node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return traversal_order

print("BFS function implemented.")
```

**Reasoning**:
Now that the BFS function is implemented, I will call it with the defined graph and a starting node ('A') to demonstrate its functionality and print the resulting traversal order, as required by the subtask.



```python
start_node = 'A'
bfs_traversal = bfs(graph, start_node)

print(f"BFS traversal starting from node '{start_node}': {bfs_traversal}")
```

## Implement and Demonstrate DFS

### Subtask:
Implement the Depth-First Search (DFS) algorithm and demonstrate its traversal on the same sample graph used for BFS.


**Reasoning**:
I will implement the DFS algorithm as a Python function, using recursion, as instructed to track visited nodes and the traversal order.



```python
def dfs(graph, start_node, visited=None, traversal_order=None):
    if visited is None:
        visited = set()
    if traversal_order is None:
        traversal_order = []

    visited.add(start_node)
    traversal_order.append(start_node)

    for neighbor in graph[start_node]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited, traversal_order)

    return traversal_order

print("DFS function implemented.")
```

**Reasoning**:
Now that the DFS function is implemented, I will call it with the previously defined graph and starting node ('A') to demonstrate its functionality and print the resulting traversal order, as required by the subtask.



```python
start_node = 'A'
dfs_traversal = dfs(graph, start_node)

print(f"DFS traversal starting from node '{start_node}': {dfs_traversal}")
```

## Implement A* Algorithm

### Subtask:
Implement the A* search algorithm, including its heuristic function, and demonstrate its pathfinding capabilities on a simple grid or graph example. This step will show the path found by A* and its cost.


**Reasoning**:
To begin implementing the A* algorithm, I need to define the search space. According to the instructions, I will create a grid using a 2D list, marking the start ('S'), goal ('G'), obstacles ('#'), and traversable paths ('.').

**Reasoning**:
Now that the reasoning for defining the grid is stated, I will create a code block to define the grid with a start, goal, and obstacles, and then print it for verification, as required by the first instruction of the A* subtask.



```python
grid = [
    ['.', '.', '.', '#', '.', '.', '.'],
    ['.', '#', '.', '.', '.', '#', '.'],
    ['.', '#', '.', '#', '.', '.', '.'],
    ['.', '.', '.', '#', 'S', '#', '.'],
    ['#', '.', '.', '#', '.', 'G', '.'],
    ['.', '.', '.', '.', '.', '.', '.']
]

start_coords = (3, 4) # (row, col) for 'S'
goal_coords = (4, 5)  # (row, col) for 'G'

print("Defined Grid (Search Space):")
for row in grid:
    print(' '.join(row))
print(f"Start Node: {start_coords}")
print(f"Goal Node: {goal_coords}")
```

**Reasoning**:
As per the instructions, the next step for implementing the A* algorithm is to define a heuristic function. I will implement the Manhattan distance heuristic, which is suitable for grid-based pathfinding, taking current and goal coordinates as input.



```python
def manhattan_distance(coord1, coord2):
    return abs(coord1[0] - coord2[0]) + abs(coord1[1] - coord2[1])

print("Manhattan distance heuristic function implemented.")
```

**Reasoning**:
Now that the grid and heuristic function are defined, I will implement the A* search algorithm itself. This involves using a priority queue to manage nodes to visit, tracking costs (`g_score`), estimated total costs (`f_score`), and the path taken to reconstruct the optimal path.



```python
import heapq

def a_star_search(grid, start, goal, heuristic):
    rows, cols = len(grid), len(grid[0])

    # Check if start or goal are obstacles
    if grid[start[0]][start[1]] == '#' or grid[goal[0]][goal[1]] == '#':
        return None, "Start or goal is an obstacle."

    # Priority queue stores (f_score, g_score, current_node, path)
    # f_score is the total estimated cost, g_score is the cost from start to current
    # path is a list of coordinates leading to current_node
    open_set = []
    heapq.heappush(open_set, (0 + heuristic(start, goal), 0, start, [start]))

    # g_score from start to current node
    g_score = {start: 0}

    # Nodes that have already been evaluated
    closed_set = set()

    while open_set:
        f, current_g, current_node, path = heapq.heappop(open_set)

        if current_node == goal:
            return path, current_g

        if current_node in closed_set:
            continue

        closed_set.add(current_node)

        # Define possible movements (up, down, left, right)
        for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            neighbor = (current_node[0] + dr, current_node[1] + dc)

            # Check if neighbor is within grid bounds
            if 0 <= neighbor[0] < rows and 0 <= neighbor[1] < cols:
                # Check if neighbor is not an obstacle
                if grid[neighbor[0]][neighbor[1]] != '#':
                    tentative_g_score = current_g + 1 # Cost to move to neighbor is 1

                    if tentative_g_score < g_score.get(neighbor, float('inf')):
                        g_score[neighbor] = tentative_g_score
                        f_score = tentative_g_score + heuristic(neighbor, goal)
                        heapq.heappush(open_set, (f_score, tentative_g_score, neighbor, path + [neighbor]))

    return None, "No path found."

print("A* search algorithm implemented.")
```

**Reasoning**:
Now that the A* search function is implemented, I will call it with the previously defined grid, start coordinates, goal coordinates, and the Manhattan distance heuristic. Then, I will print the found path and its total cost, as required by the subtask.



```python
path, cost = a_star_search(grid, start_coords, goal_coords, manhattan_distance)

if path:
    print(f"Found path: {path}")
    print(f"Path cost: {cost}")

    # Visualize the path on the grid
    grid_with_path = [list(row) for row in grid] # Create a mutable copy
    for r, c in path:
        if (r, c) != start_coords and (r, c) != goal_coords:
            grid_with_path[r][c] = '*'

    print("\nGrid with Path:")
    for row in grid_with_path:
        print(' '.join(row))
else:
    print(f"No path found. Reason: {cost}")
```

## Implement Minimax Algorithm

### Subtask:
Implement the basic Minimax algorithm for a two-player deterministic game, such as a simplified game tree.


#### **Reasoning**:
To begin implementing the Minimax algorithm, I need to define a game tree as a nested dictionary structure, where leaf nodes have associated scores (utilities), as specified by the first instruction.

**Reasoning**:
Following the reasoning, I will now define the game tree using a nested dictionary structure, where leaf nodes have utility scores, and print it to demonstrate the game's setup.



```python
game_tree = {
    'A': { # Max player's turn
        'B': {
            'D': 3,
            'E': 12
        },
        'C': {
            'F': 8,
            'G': 2
        }
    }
}

print("Defined Game Tree:")
def print_tree(node, indent=0):
    if isinstance(node, dict):
        for key, value in node.items():
            print('  ' * indent + str(key))
            print_tree(value, indent + 1)
    else:
        print('  ' * indent + str(node))

print_tree(game_tree)
```

**Reasoning**:
Now that the game tree is defined, I will implement the Minimax algorithm function. This function will recursively evaluate the game tree to determine the optimal value for the current player, as required by the subtask instructions.



```python
def minimax(node, depth, maximizing_player):
    if isinstance(node, (int, float)): # If it's a leaf node (utility score)
        return node

    if maximizing_player:
        max_eval = float('-inf')
        for child in node.values():
            eval = minimax(child, depth + 1, False)
            max_eval = max(max_eval, eval)
        return max_eval
    else: # Minimizing player
        min_eval = float('inf')
        for child in node.values():
            eval = minimax(child, depth + 1, True)
            min_eval = min(min_eval, eval)
        return min_eval

print("Minimax algorithm function implemented.")
```

**Reasoning**:
Now that the Minimax algorithm is implemented, I will call it on the root of the `game_tree` to find the optimal value. I will also extend the implementation to determine the optimal move for the starting player, as required by the subtask instructions, and print both the optimal value and the optimal move.



```python
def minimax_with_move(node, depth, maximizing_player):
    if isinstance(node, (int, float)): # If it's a leaf node (utility score)
        return node, None # Return score and no move for leaf node

    if maximizing_player:
        max_eval = float('-inf')
        best_move = None
        for move, child in node.items():
            eval, _ = minimax_with_move(child, depth + 1, False)
            if eval > max_eval:
                max_eval = eval
                best_move = move
        return max_eval, best_move
    else: # Minimizing player
        min_eval = float('inf')
        best_move = None
        for move, child in node.items():
            eval, _ = minimax_with_move(child, depth + 1, True)
            if eval < min_eval:
                min_eval = eval
                best_move = move
        return min_eval, best_move

# Call Minimax from the root node 'A', assuming 'A' is a maximizing player's turn
optimal_value, optimal_move = minimax_with_move(game_tree['A'], 0, True)

print(f"Optimal value for the maximizing player (from root 'A'): {optimal_value}")
print(f"Optimal first move for the maximizing player: {optimal_move}")
```
