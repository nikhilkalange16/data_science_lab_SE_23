### 1. Factorial of a given number

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

num = 5
print(f"The factorial of {num} is {factorial(num)}")
```

### 2. Display multiplication table for a given number

```python
def display_table(n):
    print(f"Multiplication Table for {n}:")
    for i in range(1, 11):
        print(f"{n} x {i} = {n*i}")

num = 7
display_table(num)
```

### 3. Identify if the given number is odd or even

```python
def check_odd_even(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

num1 = 4
num2 = 7
print(f"The number {num1} is {check_odd_even(num1)}.")
print(f"The number {num2} is {check_odd_even(num2)}.")
```

### 4. Identify if the entered number is palindrome or not

```python
def check_palindrome(n):
    # Convert number to string to easily reverse it
    s = str(n)
    if s == s[::-1]:
        return True
    else:
        return False

num1 = 121
num2 = 123
print(f"Is {num1} a palindrome? {check_palindrome(num1)}")
print(f"Is {num2} a palindrome? {check_palindrome(num2)}")
```

### 5. Develop an Expert System that provides simple decision-making

Here's a simple rule-based expert system to identify an animal based on a few characteristics.

```python
def animal_expert_system():
    print("Welcome to the Simple Animal Identification Expert System!")
    print("Please answer the following questions with 'yes' or 'no'.")

    has_fur = input("Does the animal have fur? ").lower()
    lays_eggs = input("Does the animal lay eggs? ").lower()
    has_wings = input("Does the animal have wings? ").lower()

    if has_fur == 'yes' and lays_eggs == 'no' and has_wings == 'no':
        print("It might be a Mammal (e.g., Cat, Dog).")
    elif has_fur == 'no' and lays_eggs == 'yes' and has_wings == 'yes':
        print("It might be a Bird.")
    elif has_fur == 'no' and lays_eggs == 'yes' and has_wings == 'no':
        print("It might be a Reptile or Amphibian (e.g., Snake, Frog).")
    else:
        print("I cannot identify the animal with the given information. Please try different characteristics.")

animal_expert_system()
```
